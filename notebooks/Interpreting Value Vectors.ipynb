# Inspecting the value vectors' projections in vocabulary space
We project 36 value vectors from actual sources of toxicity on vocabulary space and get the top 5 promoted words. (To reproduce the complete table, change `topk` in `get_value_vector_projections` to 30.)

```python
import warnings

warnings.filterwarnings("ignore")

import os.path as osp
import pickle

from rich.pretty import pprint

from xg import ASSET_DIR, WEIGHT_DIR
from xg.interp import get_top_toxic_value_vectors, get_value_vector_projections
```

```python
with open(osp.join(ASSET_DIR, "actual_sources_of_toxicity.pkl"), "rb") as f:
    ast = pickle.load(f)
```

```python
projections = get_value_vector_projections(
    model_name="ai-forever/mGPT", columns=ast, topk=5
)
```

```python
pprint(projections)
```

## Getting the potential sources of toxicity 


```python
probe_path = osp.join(WEIGHT_DIR, "english-probe.pt")
pst = get_top_toxic_value_vectors(
    probe_path=probe_path, model_name="ai-forever/mGPT", topk=100
)
```

```python
pprint(pst)
```
